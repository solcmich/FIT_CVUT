```python
def plot_pca(dataset):
    """
    Plots observed metrics for PCA on examind components count.
    """
    p = dataset[['preprocessor', 'f1', 'precision', 'recall']].groupby('preprocessor').mean()
    labels =  ['None', 'PCA(10)', 'PCA(100)', 'PCA(20)', 'PCA(40)','PCA(5)', 'PCA(60)', 'PCA(80)']
    x = np.arange(len(labels))
    width = 0.20
    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, p['f1'].values, width, label='f1')
    rects2 = ax.bar(x + width/2, p['precision'].values, width, label='precision')
    rects3 = ax.bar(x + 1.5*width, p['recall'].values, width, label='recall')
    ax.set_ylabel('Scores')
    ax.set_title('Scores for normalization and PCA')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    x1,x2,y1,y2 = plt.axis()  
    plt.axis((x1,x2,0,1))
    fig.tight_layout()
    plt.show()
```

# Hyperparameter tuning
## Dataset A
### Normalization, PCA(40), Random forest, 5% prior

```python
if DATA_AVAILABLE:
    X, y = split(charged_off)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12, stratify=y)
```

```python
clf_r_forrest = RandomForestClassifier(random_state=12, bootstrap=False, criterion='entropy', max_depth=10, min_samples_leaf=4, min_samples_split=2, n_estimators=200)
p = Pipeline(steps=[('normalize', preprocessing.Normalizer()), ('pca', PCA(40)), ('f', clf_r_forrest)])
if DATA_AVAILABLE:
    t = tune_params(p, forest_grid , 0.05, X_train, y_train, 'results/PCA/RF_5_A.csv', n_iter=100)
else:
    display(pd.read_csv('results/PCA/RF_5_A.csv').sort_values('prec', ascending=False).head(1))
```
