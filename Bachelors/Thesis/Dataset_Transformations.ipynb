```python
def plot_ptransformations(dataset):
    p = dataset[['preprocessor', 'f1', 'precision', 'recall']].groupby('preprocessor').mean()
    labels =  ['None', 'Binning', 'Abs', 'Norm', 'Robust', 'Scaler']
    x = np.arange(len(labels))
    width = 0.20
    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, p['f1'].values, width, label='f1')
    rects2 = ax.bar(x + width/2, p['precision'].values, width, label='precision')
    rects3 = ax.bar(x + 1.5*width, p['recall'].values, width, label='recall')
    ax.set_ylabel('Scores')
    ax.set_title('Scores for transformations')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    x1,x2,y1,y2 = plt.axis()  
    plt.axis((x1,x2,0,1))
    fig.tight_layout()
    plt.show()
```

```python
plot_ptransformations(pd.read_csv('experiments/Transformations1_Marked.csv').fillna(0))
```

```python
plot_ptransformations(pd.read_csv('experiments/Transformations1_Charged_Off.csv').fillna(0))
```

# Model tuning
## Dataset A

```python
if DATA_AVAILABLE:
    X, y = split(charged_off)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12, stratify=y)
```

### RF, Normalization, 25% prior - F1 score

```python
clf_r_forrest = RandomForestClassifier(random_state=12, class_weight='balanced', criterion='entropy', max_depth=10, max_features='sqrt', min_samples_leaf=4, min_samples_split=2, n_estimators=100)
p = Pipeline(steps=[('normalize', preprocessing.Normalizer()), ('f', clf_r_forrest)])
if DATA_AVAILABLE:
    t = tune_params(p, forest_grid , 0.25, X_train, y_train, 'results/Dataset_Transformations/RF_Norm_25.csv', n_iter=100)
else:
    display(pd.read_csv('results/Dataset_Transformations/RF_Norm_25.csv').sort_values('f1', ascending=False).head(1))
```

```python
test_model(X_train, X_test, y_train, y_test, p, 0.25)
```

```python
ann_prec = MLPClassifier(random_state=12, activation='tanh', alpha=0.01, hidden_layer_sizes=(50, 5), learning_rate='adaptive', solver='sgd')
ann_f = MLPClassifier(random_state=12, activation='logistic', alpha=0.0001, hidden_layer_sizes=(50, 50, 50), learning_rate='constant', solver='adam')
p1 = Pipeline(steps=[('kbins', preprocessing.KBinsDiscretizer(encode='ordinal')), ('f', ann_prec)])
p2 = Pipeline(steps=[('kbins', preprocessing.KBinsDiscretizer(encode='ordinal')), ('f', ann_f)])
if DATA_AVAILABLE:
    t = tune_params(p, ann_space , 0.05, X_train, y_train, 'results/Dataset_Transformations/ANN_KBins_5.csv', n_iter=100)
display(pd.read_csv('results/Dataset_Transformations/ANN_KBins_5.csv').sort_values('prec', ascending=False).head(1))
display(pd.read_csv('results/Dataset_Transformations/ANN_KBins_5.csv').sort_values('f1', ascending=False).head(1))  
```
